```python
# Dependencies
import os
import pandas as pd
import glob
```

```python
# Getting a list of filenames 
filenames=glob.glob("Data/TripData-2017/*tripdata.csv")
num_files=len(filenames)
print(num_files)

combined_df = pd.DataFrame()
for csv_file in filenames:
       
    df = pd.read_csv(csv_file, header=[0])
    # df.columns       
        
    df.set_axis(['Trip Duration', 'Start Time', 'Stop Time', 'Start Station ID', 'Start Station Name', 'Start Station Latitude',
                      'Start Station Longitude', 'End Station ID', 'End Station Name', 'End Station Latitude', 'End Station Longitude', 
                      'Bike ID', 'User Type', 'Birth Year', 'Gender'], axis=1, inplace = True)
        
    print (df.head(5))
        
    combined_df=pd.concat([combined_df, df])    
```
